```python
import random

nama_file = "data.txt"

with open(nama_file, "w") as file:
    for _ in range(10):
        angka = random.randint(1, 100)
        file.write(f"{angka}\n")

print(f"File '{nama_file}' berhasil dibuat!")
```

```python
print("--- Hasil Analisis ---")

try:
    with open("data.txt", "r") as file:
        for baris in file:
            angka = int(baris.strip())

            if angka % 2 == 0:
                status = "Genap"
            else:
                status = "Ganjil"

            print(f"Angka {angka} adalah {status}")

except FileNotFoundError:
    print("File tidak ditemukan. Jalankan kode langkah 1 terlebih dahulu.")
```
